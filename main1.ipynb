```python
from collections import Counter

def canConstruct(string1, string2):
    
    string1_counter = Counter(string1)
    string2_counter = Counter(string2)
    
    for char,count in string1_counter.items():
        if string2_counter[char] < count:
            return False
        
    return True

string1 = "aabbcd"
string2 = "aabbcdddd"

print(canConstruct(string1,string2))
```

```python
def word_pattern(pattern: str, s: str) -> bool:
    words = s.split()
    
    if len(pattern) != len(words):
        return False

    char_to_word = {}
    word_to_char = {}
    
    for char,word in zip(pattern,words):
        
        if char in char_to_word:
            if char_to_word[char] != word:
                return False
        else:
            char_to_word[char] = word
        
        if word in word_to_char:
            if word_to_char[word] != char:
                return False
        else:
            word_to_char[word] = char
    
    print(char_to_word,word_to_char)
    return True

string = "dog cat cat dog"
pattern = "abba"
print(word_pattern(pattern,string))
```

```python
def sort_by_binary_ones(arr):
    # Define a key function that returns a tuple (number of ones, original number)
    # This tuple is used to sort the array
    def sort_key(x):
        return (bin(x).count('1'), x)

    # Sort the array using the defined key function
    sorted_arr = sorted(arr, key=sort_key)
    return sorted_arr

# Example usage
arr = [7, 8, 6, 5]
sorted_arr = sort_by_binary_ones(arr)
print("Sorted array:", sorted_arr)
```

```python
def lonelyinteger(a):

    count_dict = {}
    for item in a:
        if item in count_dict:
            count_dict[item] += 1
        else:
            count_dict[item] = 1
    
    # Find the item that appears only once
    for key, value in count_dict.items():
        if value == 1:
            return key


a = [1, 2, 3, 2, 3]
print(lonelyinteger(a))  
```

```python
from collections import Counter

string1 = "anagram"
string2 = "nagaram"

def is_anagram(string1,string2):
    
    if len(string1) != len(string2):
        return False
    
    else:
        return Counter(string1) == Counter(string2)
    

print(is_anagram(string1,string2))
```

```python
nums = [0,1,2,4,5,7]
results = []
start = nums[0]
for i in range(len(nums)-1):
    if nums[i] + 1 != nums[i+1]:
        end = nums[i]
        if start == end:
            results.append(str(start))
        else:
            results.append([f"{start}->{end}"])
        start = nums[i+1]
    
if start == nums[-1]:
    results.append(str(start))

else:
    results.append(f"{start}->{nums[-1]}")

print(results)
```
